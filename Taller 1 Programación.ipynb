```python
# 10. Leer una matriz e imprimir el vector que aparece cuando la matriz se 
#desenreda en forma de espiral que gira en el sentido horario a partir de la 
#esquina inferior derecha.        

t =('\n************CONVERTIDOR DE MATRIZ A VECTOR EN FORMA ESPIRal**********\
    ******')
print(t.upper())
            
import numpy as np
matriz = []

filas = int(input('Cantidad de filas: '))
columnas = int(input('Cantidad de columnas: ')) 

for k in range(filas):
    matriz.append([])          

    for j in range(columnas):
        matriz[k].append(int(input('Elemento %d,%d : '% ( k, j  ))))
        
x = np.array(matriz)
print('~~~~~~~~~~~~~\n',x)

elementos = filas * columnas 
vec = []
c_final = columnas - 1
c_inicio = 0
f_final = filas - 1
f_inicio = 0
aux = 0
   
    
while aux < elementos:
    
    for k in range(c_final, c_inicio, -1):
        ex = matriz[f_final][k]
        vec.append(ex)
        aux += 1
    for k in range(f_final, f_inicio, -1):
        ex = matriz[k][c_inicio]
        vec.append(ex)
        aux += 1    
    for k in range(c_inicio, c_final):
        ex = matriz[f_inicio][k]
        vec.append(ex)
        aux += 1
    for k in range(f_inicio, f_final):
        ex = matriz[k][c_final]
        vec.append(ex)
        aux += 1
        
    c_final -= 1
    c_inicio += 1
    f_final -= 1
    f_inicio += 1    
    
print('\n~~~~~~~~~~~~~~~~~~~~~')
print(vec)
print('\n********FIN*************')   

        
```

```python
# 14. Diseñe una función que proponga sumas de números positivos (dos números 
#entre 1 y 100) al usuario y compruebe la respuesta. El programa continuará 
#hasta que se acierten cinco sumas. (Tip:para generar los números enteros 
#aleatorios, existe la función random.randint(a,b) de la librería random).

h =('\n*************SUMA DE NÚMEROS POSITIVOS****************')
print(h.upper())    
    
import random
def suma():  
    c = 0
    while c < 5:
        a = random.randrange(1, 101)#Número a aleatorio entre un rango
        b = random.randrange(1, 101)
        ejercicio = int(input(f'{a} + {b} = '))
        if ejercicio == a + b:
            c += 1
            print('La respuesta es correcta!')
        else:
            print("¡Respuesta incorrecta!", '\nVuelva a Intentar')    
            
    print("\nPrograma terminado.") 
    print('FIN')       
    
suma()
```
